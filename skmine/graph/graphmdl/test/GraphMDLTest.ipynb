In this example, we will apply the GraphMDL+ algorithm on a graph.
GraphMDL+ uses the Minimum Description Length principle to facilitate the exploration of models on graph data, and provide an easy-to-use result.


```python
import networkx as nx
from skmine.graph.graphmdl.graph_mdl import GraphMDL
from skmine.graph.graphmdl import utils
```

GraphMDL+ is used only for graphs.
So, first, we create a graph with the networkX python library.
Note : Graph edges must be labeled

```python
data_graph = nx.DiGraph() # Create new oriented graph
#Add eight nodes (from 1 to 8) to the graph
data_graph.add_nodes_from(range(1,9))
#Add each node label
data_graph.nodes[1]['label'] = 'y'
data_graph.nodes[2]['label'] = 'x'
data_graph.nodes[3]['label'] = 'z'
data_graph.nodes[4]['label'] = 'x'
data_graph.nodes[5]['label'] = 'z'
data_graph.nodes[6]['label'] = 'x'
data_graph.nodes[7]['label'] = 'z'
data_graph.nodes[8]['label'] = 'w','x'

#Add all edges and their label
data_graph.add_edge(2,1,label='a')
data_graph.add_edge(4,1,label='a')
data_graph.add_edge(6,1,label='a')
data_graph.add_edge(6,8,label='a')
data_graph.add_edge(8,6,label='a')
data_graph.add_edge(1,3,label='b')
data_graph.add_edge(1,5,label='b')
data_graph.add_edge(1,7,label='b')
```

NetworkX can display the graph that you have just created, see below the network graph

```python
utils.draw_graph(data_graph)
```

Note : GraphMDL+ can be run with a timeout or not.

To run GraphMDL+, we will first initialize the GraphMDL+ algorithm, and after we will fit the algorithm with a data graph

Now we’re going to run GraphMDL+ with the previous data graph and an execution timeout.

The timeout is the total of seconds of the execution, and it’s optional.

```python
graphmdl = GraphMDL()
graphmdl.fit(data_graph, timeout=0.50)
```

After execution, we can retrieve the found patterns and treat them.

For example, we will display below the structure of the models found with the display of networkX graphs.

Note: Patterns are also networkX graphs.

```python
for p in graphmdl.patterns():
    utils.draw_pattern(p)
```

We can also get the final and the initial description length, and treat them.

Below an example

```python
print(f"initial description length = {graphmdl.initial_description_length()}")
print(f"Final description length = {graphmdl.description_length()}")
print(f"GraphMDL+ description length percent = {round((graphmdl.description_length()/ graphmdl.initial_description_length())*100, 2)} %")
```

The timeout in the GRAPHMDL running is optional.
